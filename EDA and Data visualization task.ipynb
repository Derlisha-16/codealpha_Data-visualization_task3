```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(r"C:\Users\DERLISHA\OneDrive\Documents\Attachments\Desktop\Bookscrap_excel.xlsx")
df
```

```python
#Data distribution using histogram
plt.figure(figsize=(10, 6))   # increase size

plt.hist(df['Price'], bins=15)

plt.title("Distribution of Product Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.xticks(rotation=45)       # rotate x labels
plt.tight_layout()            # adjust spacing
plt.show()
```

```python
#price analysis using box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Price'])
plt.title("Price Spread and Outliers")
plt.xticks(rotation=45)       # rotate x labels
plt.tight_layout()
plt.show()
```

```python
#product availability count
plt.figure()
sns.countplot(x='instock', data=df)
plt.xlabel("Stock Status")
plt.ylabel("Count")
plt.title("Product Availability Status")
plt.show()
```

```python
#Relationship between price and availability
plt.figure()
sns.barplot(x='instock',y='Price',data=df)
plt.title("Average Price vs Stock Availability")
plt.show()
```
